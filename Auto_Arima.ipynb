```python
df_data = pd.DataFrame(data2, columns = ['timestamp','open (USD)', 'high (USD)',
                                            'low (USD)', 'close (USD)','volume'])
```

```python
#Drop columns which contain only missing values
df_column_drop = data.dropna(axis=1, how='all')
df_column_drop
```

```python
# Select the rows of where data is not NaN
df_notnull = data[data['open (USD)'].notnull()& data['high (USD)'].notnull()&data['low (USD)'].notnull()&data['close (USD)']]
df_notnull
```

```python
train_data=df_data[df_data.iloc[:,0]<'2018-01-01']
train_data=train_data.iloc[::-1]
train_data
```

```python
test_data=df_data[df_data.iloc[:,0]>='2018-01-01']
test_data=test_data.iloc[::-1]
test_data
```

```python
  train_coin_price = train_data.iloc[:, 1:2].values
  train_coin_price
```

```python
import numpy as np
train_coin_price=np.flipud(train_coin_price)
```

```python
real_coin_price = test_data.iloc[:, 1:2].values
real_coin_price
```

```python
import matplotlib.pyplot as plt

from pyramid.arima import auto_arima
model = auto_arima(train_coin_price, trace=True, error_action='ignore', suppress_warnings=True)
model.fit(train_coin_price)
```

```python
import pandas as pd
total_open = pd.concat((train_data['open (USD)'], test_data['open (USD)']), axis = 0)
```

```python
test_coin_price = total_open[len(total_open) - len(test_data) - 60:].values
test_coin_price = test_coin_price.reshape(-1,1)
```

```python
#test_coin_price = sc.transform(test_coin_price)
X_test = []
for i in range(60, 500):
    X_test.append(test_coin_price[i-60:i, 0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
predicted_coin_price = model.predict(n_periods=len(X_test))
#predicted_coin_price = sc.inverse_transform(predicted_coin_price)
```

```python
plt.plot(real_coin_price, label='real coin price')
#plt.plot(real_coin_price, label='real_coin_price')
plt.plot(predicted_coin_price, label='predicted coin price')
plt.legend()
plt.title('AutoArima Model')
```

```python
def classifcation(df):
    list_y =[]
    temp = df
    for i in range(len(temp)-1):
        if temp[i+1] > temp[i]:
            y = 1
        else:
            y = -1
            
        list_y.append(y)
    ydf = pd.DataFrame(list_y)
    return ydf
```

```python
real_coin_price_2 = classifcation(real_coin_price[0:440]) #[0:67]
predicted_coin_price_2 =classifcation(predicted_coin_price)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(real_coin_price_2, predicted_coin_price_2)
print('Accuracy for Autoarima Model',score)
```

```python
from sklearn.metrics import f1_score
f1_score(real_coin_price_2, predicted_coin_price_2)
```
